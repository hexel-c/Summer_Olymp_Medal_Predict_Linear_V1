## **Cleaning Data**

Check for missing data

```python
teams[teams.isnull().any(axis=1)]
```

Drop Missing Data

```python
teams=teams.dropna()

teams
```

## **Making The Model**

Split Data

```python
train = teams[teams["year"] < 2012].copy()

test = teams[teams["year"] >= 2012].copy()
```

Check rows and columns of training and test sets

```python
train.shape
```

```python
test.shape
```

## **Training The Model**

```python
from sklearn.linear_model import LinearRegression

reg = LinearRegression() # This is the trained model
```

Creating Predictors

```python
predictors = ["athletes", "age", "height", "weight", "prev_medals"]
```

Create Target

```python
target = "medals"
```

## **Fit the model**

```python
reg.fit(train[predictors], train["medals"])
```

## **Making Predictions**

```python
predictions = reg.predict(test[predictors])

predictions
```

**re-scale predictions to make it more readable and clearer**

Assign predictions to a column in our test dataframe

```python
test["predictions"] = predictions

test
```

Replace prediction value <0 to 0 (remove neagtives)

```python
test.loc[test["predictions"] < 0, "predictions"] = 0
```

Round up the data

```python
test["predictions"] = test["predictions"].round()
```

```python
test
```

**Check for mean absolute error**

```python
from sklearn.metrics import mean_absolute_error

error = mean_absolute_error(test["medals"], test["predictions"])
```

```python
error
```

Check if our mean prediction error is a good or bad error

```python
teams.describe()["medals"]
```

## **Check error by country**

```python
errors = (test["medals"] - test["predictions"]).abs()

errors
```

Group by team

```python
error_by_team  = errors.groupby(test["team"]).mean()

error_by_team
```

Check how many medals each country earned on average

```python
medals_by_team = test["medals"].groupby(test["team"]).mean()
```

Find the ratio between our errors

```python
error_ratio = error_by_team / medals_by_team

error_ratio
```

Take only values that are not missing from the error ratio remove NaN

```python
error_ratio[~pd.isnull(error_ratio)]
```

Clean up infinite values

```python
import numpy as np
error_ratio = error_ratio[np.isfinite(error_ratio)]
```

```python
error_ratio
```

Create a histogram for the error ratio

```python
error_ratio.plot.hist()
```

**Check which country tends to get a lot of medal from the olympics**

For example our error is very low for countries who tend to send more athletes to the olympics

```python
error_ratio.sort_values()
```
